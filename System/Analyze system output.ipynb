# Import the data

```python
# Import necessary packages
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib
from matplotlib.dates import DateFormatter
import datetime
import pandas as pd
import matplotlib.patches as mpatches

# Import the data and make sure that:
# 'tweet_hashtags' column is a list
oneminute_windows = pd.read_csv('Output/1 minute window/windows-1-minute-0125.csv', index_col=0)
fiveminute_windows = pd.read_csv('Output/5 minute window/windows-5-minute-0125.csv', index_col=0)
```

```python
oneminute_windows["start_window"] = pd.to_datetime(oneminute_windows["start_window"])
fiveminute_windows["start_window"] = pd.to_datetime(fiveminute_windows["start_window"])
```

```python
oneminute_windows = oneminute_windows[oneminute_windows["start_window"]<="2020-06-03 18:15:06"]
fiveminute_windows = fiveminute_windows[fiveminute_windows["start_window"]<="2020-06-03 18:15:06"]
```

# One minute timewindow

```python
import matplotlib.patches as mpatches
```

```python
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot()

ax.plot(oneminute_windows["start_window"], oneminute_windows["n_tweets"], label="All tweets")
date_form = DateFormatter("%H:%M")
ax.xaxis.set_major_formatter(date_form)

ax.set_ylabel('Number of tweets', fontsize=15)
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)

ax.set_axisbelow(True)
ax.grid(True, axis='y', color="#e5e5e5")
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

blue_patch = mpatches.Patch(color='#247ab5', label='All tweets')
orange_patch = mpatches.Patch(color='#ff9400', label='Events')
red_patch = mpatches.Patch(color='#ff2600', label='Incidents')
plt.legend(handles=[blue_patch, orange_patch, red_patch], fontsize=12)
plt.savefig('Plots/windows-1-minute-over-time.png', dpi=300)
plt.show()
```
