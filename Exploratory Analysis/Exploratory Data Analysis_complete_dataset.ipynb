```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib
from matplotlib.dates import DateFormatter
import datetime
import numpy as np
import matplotlib.patches as mpatches

# Function to count how many times a string value exists in a list
def count_values_list(df, variable):
    # Set counter (necessary for for-loop) and create empty dataframe
    i = 0
    values_df = pd.DataFrame(columns = ['value', 'count'])
    
    # Loop through each values element
    for index, values in df[variable].iteritems():
        
        # Check if values is a string (then convert to list)
        if type(values) == str:
            values = values.strip('][').split(', ') 
        
        # Check if there are any values (if list --> values available)
        if type(values) == list:
            
            # Loop through all values
            for value in values:
                
                value = value.lower()
                
                # If value is already in dataframe, add +1 to count
                if (values_df['value']==value).any() == True:
                    index_row = values_df[values_df['value']==value].index.values.astype(int)[0]
                    values_df.loc[index_row, 'count'] = values_df.loc[index_row, 'count'] + 1
                
                # Create new value in dataframe
                else:
                    values_df.loc[i,'value'] = value
                    values_df.loc[i,'count'] = 1
                    i = i+1
    return values_df
```

## Check how other user types are mentioned over time
Above, only the top 5 mentioned user types are taken into account, following the approach of Hu et al. (2012). Here, we will look for insightful patterns with the other user types.

```python
# Create dataframes for the other mentioned typs
political_party_df = pd.DataFrame()
comedian_df = pd.DataFrame()
political_org_df = pd.DataFrame()
political_act_df = pd.DataFrame()
writer_df = pd.DataFrame()
part_of_government_org_df = pd.DataFrame()
government_org_df = pd.DataFrame()
actress_df = pd.DataFrame()
musician_df = pd.DataFrame()

# Store the rows in seperate DataFrames
for index,row in data_after_3.iterrows():
    if (row["user_mentions_types"] != None) and (type(row["user_mentions_types"]) != float):
        if "political party" in row['user_mentions_types']:
            political_party_df = political_party_df.append(row)
        if "comedian" in row['user_mentions_types']:
            comedian_df = comedian_df.append(row)
        if "political organization" in row['user_mentions_types']:
            political_org_df = political_org_df.append(row)
        if "political activist" in row['user_mentions_types']:
            political_act_df = political_act_df.append(row)
        if "writer" in row['user_mentions_types']:
            writer_df = writer_df.append(row)
        if "part of government organization" in row['user_mentions_types']:
            part_of_government_org_df = part_of_government_org_df.append(row)
        if "government organization" in row['user_mentions_types']:
            government_org_df = government_org_df.append(row)
        if "actress" in row['user_mentions_types']:
            actress_df = actress_df.append(row)
        if "musician" in row['user_mentions_types']:
            musician_df = musician_df.append(row)
```

```python
# Resample all the dataframes to 15-minute timeframes
political_party_df_agg_15_min = political_party_df.resample('5Min', on='created_at').sum()
political_party_df_agg_15_min.reset_index(inplace=True)

# Resample all the dataframes to 15-minute timeframes
comedian_df_agg_15_min = comedian_df.resample('5Min', on='created_at').sum()
comedian_df_agg_15_min.reset_index(inplace=True)

# Resample all the dataframes to 15-minute timeframes
political_org_df_agg_15_min = political_org_df.resample('5Min', on='created_at').sum()
political_org_df_agg_15_min.reset_index(inplace=True)

# Resample all the dataframes to 15-minute timeframes
political_act_df_agg_15_min = political_act_df.resample('5Min', on='created_at').sum()
political_act_df_agg_15_min.reset_index(inplace=True)

# Resample all the dataframes to 15-minute timeframes
writer_df_agg_15_min = writer_df.resample('5Min', on='created_at').sum()
writer_df_agg_15_min.reset_index(inplace=True)

# Resample all the dataframes to 15-minute timeframes
part_of_government_org_df_agg_15_min = part_of_government_org_df.resample('5Min', on='created_at').sum()
part_of_government_org_df_agg_15_min.reset_index(inplace=True)

# Resample all the dataframes to 15-minute timeframes
government_org_df_agg_15_min = government_org_df.resample('5Min', on='created_at').sum()
government_org_df_agg_15_min.reset_index(inplace=True)

# Resample all the dataframes to 15-minute timeframes
actress_df_agg_15_min = actress_df.resample('5Min', on='created_at').sum()
actress_df_agg_15_min.reset_index(inplace=True)

# Resample all the dataframes to 15-minute timeframes
musician_df_agg_15_min = musician_df.resample('5Min', on='created_at').sum()
musician_df_agg_15_min.reset_index(inplace=True)
```
