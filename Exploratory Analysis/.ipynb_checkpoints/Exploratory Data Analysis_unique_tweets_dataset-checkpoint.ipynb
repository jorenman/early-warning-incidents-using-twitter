```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib
from matplotlib.dates import DateFormatter
import datetime
import pytz
from urllib.parse import urlparse
import numpy as np

# Function to count how many times a string value exists in a list
def count_domains(df, variable):
    # Set counter (necessary for for-loop) and create empty dataframe
    i = 0
    values_df = pd.DataFrame(columns = ['value', 'count'])
    
    # Loop through each values element
    for index, values in df[variable].iteritems():
        
        # Check if values is a string (then convert to list)
        if type(values) == str:
            values = values.strip('][').split(', ') 
        
        # Check if there are any values (if list --> values available)
        if type(values) == list:
            
            # Loop through all values
            for value in values:
                
                value = value.lower()
                
                # Grab the domain of the URL
                value = urlparse(value).netloc
                
                if 'www.' in value:
                    value = value.replace('www.', '')
                    
                # If value is already in dataframe, add +1 to count
                if (values_df['value']==value).any() == True:
                    index_row = values_df[values_df['value']==value].index.values.astype(int)[0]
                    values_df.loc[index_row, 'count'] = values_df.loc[index_row, 'count'] + 1
                
                # Create new value in dataframe
                else:
                    values_df.loc[i,'value'] = value
                    values_df.loc[i,'count'] = 1
                    i = i+1
    return values_df

# Function to count how many times a string value exists in a list
def count_values_list(df, variable):
    # Set counter (necessary for for-loop) and create empty dataframe
    i = 0
    values_df = pd.DataFrame(columns = ['value', 'count'])
    
    # Loop through each values element
    for index, values in df[variable].iteritems():
        
        # Check if values is a string (then convert to list)
        if type(values) == str:
            values = values.strip('][').split(', ') 
        
        # Check if there are any values (if list --> values available)
        if type(values) == list:
            
            # Loop through all values
            for value in values:
                
                value = value.lower()
                
                # If value is already in dataframe, add +1 to count
                if (values_df['value']==value).any() == True:
                    index_row = values_df[values_df['value']==value].index.values.astype(int)[0]
                    values_df.loc[index_row, 'count'] = values_df.loc[index_row, 'count'] + 1
                
                # Create new value in dataframe
                else:
                    values_df.loc[i,'value'] = value
                    values_df.loc[i,'count'] = 1
                    i = i+1
    return values_df
```

## First ten incident-related tweets over time
In this section, we will look at how the first ten incident-related tweets are mentioned over time. In order to do this, the unique tweets dataset is combined with the complete tweets dataset.

```python
# Import complete tweets dataset
complete_data = pd.read_pickle("~/Documents/Github Repository/early-warning-twitter/Processed datasets/Tweets/01-06-2020-amsterdam-demonstration.pkl")

# Only get the data from 16:50 to 18:00
complete_data = complete_data[(complete_data["created_at"]>= "2020-06-01 16:50:00")&(complete_data["created_at"]<="2020-06-01 18:00:00")]
complete_data.reset_index(inplace=True, drop=True)

# Check how many times the first incident related tweets are mentioned in the complete dataset
incident_related_mentions = pd.DataFrame(columns=["screen_name", "count"])
incident_related_mention_tweets = pd.DataFrame(columns=complete_data.columns)

# Get the first ten incident_related tweets
first_ten_incident_related = incident_related_after_3.sort_values('created_at').head(10)

# Lowercase all the screen names
first_ten_incident_related["user_screen_name"] = first_ten_incident_related["user_screen_name"].str.lower()

# Necessary for handling the for-loop
o = 0

for index,row in first_ten_incident_related.iterrows():
    screen_name = row["user_screen_name"]
    
    for i in range(len(complete_data)):
        if complete_data.loc[i, 'user_mentions'] != None:
            
            # Check if the user is mentioned in the complete dataset
            if (screen_name in complete_data.loc[i, 'user_mentions']) == True:
                
                # If the user is mentioned, add the tweet to incident_related_mention_tweets (if it is not already there)
                if (incident_related_mention_tweets['id']==complete_data.loc[i, 'id']).any() == False:
                    incident_related_mention_tweets = incident_related_mention_tweets.append(complete_data.loc[i,])
                
                # If the user is mentioned before, add 1 to the count
                if ((incident_related_mentions['screen_name']==screen_name).any() == True):
                    index_row = incident_related_mentions[incident_related_mentions['screen_name']==screen_name].index.values.astype(int)[0]
                    incident_related_mentions.loc[index_row, 'count'] = incident_related_mentions.loc[index_row, 'count'] +1
                
                # If the user is not mentioned before, add the user to the incident_related_mentions dataset
                else:
                    incident_related_mentions.loc[o, 'screen_name'] = screen_name
                    incident_related_mentions.loc[o, 'count'] = 1
                    o = o+1
```
